```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
list_years = list(range(1990, 2021))
frames = []
for year in list_years:
    df = pd.read_csv('data/games_'+str(year)+'.csv')
    del df['Unnamed: 0']
    frames.append(df)

df = pd.concat(frames, keys=list_years)
df.reset_index(drop=True, inplace=True)
df.to_csv('games.csv')
```

```python
df = pd.read_csv('games.csv')
```

```python
max_goals = max(df['home_goals'].max(), df['away_goals'].max())
array_n = [[0 for x in range(max_goals+1)] for y in range(max_goals+1)] 
sum_n = 0
list_rdiff = []
for index, row in df.iterrows():
    rdiff = row['rdiff']
    if rdiff < -100 and rdiff > -200:
        home_goals = row['home_goals']
        away_goals = row['away_goals']
        array_n[home_goals][away_goals] += 1
        sum_n += 1
        list_rdiff.append(rdiff)
        print("Index {}, rdiff {}, Result {} - {}".format(index, rdiff, home_goals, away_goals))
```

```python
fig, ax = plt.subplots()
ax.pcolor(array_n)
for i in range(max_goals+1):
    for j in range(max_goals+1):
        n = array_n[i][j]
        ax.annotate(str(n), xy=(j+0.5,i+0.45))
ax.set_xlim([0, 7])
ax.set_ylim([0, 7])
```
